```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis, color='green')
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot 
plt.barh(x_axis, y_axis, color="m", label='Boston')
plt.gca().invert_yaxis()
# Create labels for the x and y axis
plt.ylabel("Fare($)")
plt.xlabel("Date")

# Create a title 
plt.title("Pyber Fare by Month")

# Add legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="k")
```

```python
# Create the Scatter plot
plt.scatter(y_axis, x_axis, c= "r", label= "Chicago")
plt.gca().invert_yaxis()

# Add axis labels
plt.xlabel('Fare($)')
plt.ylabel('Date')

# Add x axis limit
plt.xlim(15, 50)

# Add legend
plt.legend()
```

```python
# make y axis values larger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

# Create Bubble Chart
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create Scatter plot with Object Oriented method
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 3 for i in y_axis], edgecolors= "k", alpha=.8, c="#87ceeb", label="Boston", linewidths= 2)
# Invert y axis data
plt.gca().invert_yaxis()

# Add title
plt.title("PyBer Fare by Month")

# Add x and y label
plt.xlabel( "Fare($)")
plt.ylabel("Date")

# Add x axis limit
plt.xlim(0,50)



# Add Legend
plt.legend()




```

```python
# Add a pie chart 
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Add percentage for each month
plt.subplots(figsize=(8, 8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Add x_axis, y_axis and yerr=stdev to errorbar
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add Error Bars to a Bar Chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add Error bars to Bar Chart
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
